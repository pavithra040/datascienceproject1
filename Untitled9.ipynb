```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;
df=pd.read_csv("/content/drive/MyDrive/heart.csv")
df.head()
```

```python

df.target.value_counts()
```

```python
corr=df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True,)
plt.show()
```

```python
sns.countplot(x=df.target,hue='sex',data=df,palette=['green','red'])
plt.xticks([0,1],labels=['Not Disease','Disease'])
plt.title("Target Distribution")
```

```python
print(df.target)
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['green','red'])
plt.legend(["Haven't disease","have disease"])
plt.title("Heart disease frequency for age")
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.distplot(df.age[df["target"]==0],bins=30,color="pink",label='not disease')
sns.distplot(df.age[df["target"]==1],bins=30,color="skyblue",label='disease')
plt.legend()
plt.title("Heart disease distribution for age")
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
```
